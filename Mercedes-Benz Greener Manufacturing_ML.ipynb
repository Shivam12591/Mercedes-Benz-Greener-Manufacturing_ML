### 1. removing those variable(s), In case of any column(s), the variance is equal to zero.

```python
# First removing the Columns that is not in use 
benz_train_df.drop("ID", axis=1, inplace=True)
```

```python
benz_train_df.var()
```

```python
Col_Zero_Var = benz_train_df.var()[benz_train_df.var()==0].index.values
Col_Zero_Var
```

```python
benz_train_df = benz_train_df.drop(Col_Zero_Var, axis=1)
```

```python
benz_train_df.head()
```

### 2. Checking the null and unique values for train set.

```python
# checking The number of null values
benz_train_df.isnull().sum().values
```

```python
# Now Checking the Number of unique values in each rows 
pd.DataFrame(benz_train_df.nunique(), benz_train_df.nunique().index.values)
```

### 3. Appling the Label Encoder 

```python
# Before appliyng the label encoder first need to check the object variables
benz_train_obj = benz_train_df.select_dtypes(include=object)
benz_train_obj
```

```python
# Importing the label encoder 
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
```

```python
benz_train_df["X0"] = le.fit_transform(benz_train_df["X0"])
benz_train_df["X1"] = le.fit_transform(benz_train_df["X1"])
benz_train_df["X2"] = le.fit_transform(benz_train_df["X2"])
benz_train_df["X3"] = le.fit_transform(benz_train_df["X3"])
benz_train_df["X4"] = le.fit_transform(benz_train_df["X4"])
benz_train_df["X5"] = le.fit_transform(benz_train_df["X5"])
benz_train_df["X6"] = le.fit_transform(benz_train_df["X6"])
benz_train_df["X8"] = le.fit_transform(benz_train_df["X8"])
```

```python
benz_train_df["X0"]  # category has been converted into the num and datatype is int now.
```

```python
benz_train_df
```

### 4. Performing Principal component analysis(PCA)

```python
x_benz_train_df = benz_train_df.drop('y', axis=1)
y_benz_train_df = benz_train_df['y']
```

```python
# Importing the PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=25)
```

```python
x_benz_train_df = pca.fit_transform(x_benz_train_df)
```

```python
pca.explained_variance_ratio_
```
